```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 3, 5, 7, 11]
}

# Create line plot
sns.lineplot(x='x', y='y', data=data)

# Add title and labels
plt.title('Basic Line Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```

```python
data = {
    'categories': ['A', 'B', 'C', 'D'],
    'values': [4, 7, 1, 8]
}

# Create bar plot
sns.barplot(x='categories', y='values', data=data)

# Add title and labels
plt.title('Basic Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.show()
```

```python
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [8, 6, 5, 7, 4]
}

# Create scatter plot
sns.scatterplot(x='x', y='y', data=data)

# Add title and labels
plt.title('Basic Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show plot
plt.show()
```

```python
data = {
    'category': ['A', 'A', 'A', 'B', 'B', 'B'],
    'value': [4, 5, 6, 2, 3, 1]
}

# Create box plot
sns.boxplot(x='category', y='value', data=data)

# Add title and labels
plt.title('Basic Box Plot Example')
plt.xlabel('Category')
plt.ylabel('Value')

# Show plot
plt.show()
```
